```python
#Standart imports
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
%matplotlib inline
```

```python
#PyTorch imports
import torch  
import torchvision
import torchxrayvision as xrv
import pylab
import torch.optim as optim
import torch.nn as nn
```

```python
transform = torchvision.transforms.Compose([xrv.datasets.XRayCenterCrop(),xrv.datasets.XRayResizer(224)])
idk = xrv.datasets.COVID19_Dataset(imgpath='D:\Licon\Xray\covid-chestxray-dataset\images',csvpath='D:\Licon\Xray\covid-chestxray-dataset\metadata.csv',transform=transform)
```

```python
df = idk.csv.drop(['offset','date', 'location', 'folder', 'filename', 'doi', 'url', 'license', 'clinical_notes', 'other_notes', 'Unnamed: 28', 'survival', 'intubated', 'intubation_present', 'went_icu', 'in_icu', 'needed_supplemental_O2', 'extubated', 'temperature', 'pO2_saturation', 'leukocyte_count', 'neutrophil_count', 'lymphocyte_count'], axis = 1)
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
len_dataset=len(idk)
n_train=int(0.9*len_dataset)
n_test=int(0.1*len_dataset)+1
print(len_dataset,n_train,n_test)
```

```python
train_dataset,test_dataset=torch.utils.data.random_split(idk, [n_train,n_test])
dataset={'train':train_dataset,'test':test_dataset}
dataloaders = {x : torch.utils.data.DataLoader(dataset[x], batch_size=8,shuffle=False, num_workers=4) for x in ['train','test'] }
```
